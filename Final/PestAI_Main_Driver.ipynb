```python
# get webmap
from arcgis.gis import GIS
from arcgis.geocoding import geocode
from IPython.display import display

from arcgis.features import FeatureLayerCollection

our_gis = GIS(username="Ming_Wang_LearnArcGIS", password="pestai666")
our_webmap = our_gis.content.search("Pest.ai_background", item_type="Web Map")
our_webmap[0]
```

```python
# get feature layer
itemId="f8fffef0bd0c488ba0ab428ff605a9a4"
search_result = our_gis.content.search('title:Inspected GSOB Trees 0')
search_result = our_gis.content.get(itemId)
feature_layer = search_result.layers[0]
```

```python
# query current featureSet
current_fset = feature_layer.query()
print(current_fset)
print(len(current_fset))
```

```python
# new data from csv
import pandas as pd
file_name = 'prediction.csv'
df = pd.read_csv(file_name)
df.head()
```

```python
# new featureSet
from arcgis.features import GeoAccessor
from arcgis.features import FeatureSet
ga = GeoAccessor.from_xy(df, 'x', 'y')
print(ga[:5])
new_fset = FeatureSet.from_dataframe(ga)
print(new_fset)
print(len(new_fset))
```

```python
# delete current featureSet & add new featureSet
feature_layer.edit_features(adds=new_fset, deletes=current_fset)
```

```python
# check whether the update works
updated_fset = feature_layer.query()
print(updated_fset)
print(len(updated_fset))
```
